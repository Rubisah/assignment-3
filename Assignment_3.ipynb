```python
disease_data['BMI'].fillna(disease_data['BMI'].mean(), inplace=True)
disease_data['PhysicalHealth'].fillna(disease_data['PhysicalHealth'].median(), inplace=True)
disease_data['MentalHealth'].fillna(disease_data['MentalHealth'].median(), inplace=True)
disease_data['SleepTime'].fillna(disease_data['SleepTime'].mean(), inplace=True)
```

```python
disease_data.isna().sum()
```

```python
disease_data['Smoking'].fillna('No', inplace=True)
disease_data['AlcoholDrinking'].fillna('No', inplace=True)
disease_data['Stroke'].fillna('No', inplace=True)
disease_data['DiffWalking'].fillna('No', inplace=True)
disease_data['Diabetic'].fillna('No', inplace=True)
disease_data['PhysicalActivity'].fillna('Yes', inplace=True)
disease_data['GenHealth'].fillna('Very good', inplace=True)
disease_data['KidneyDisease'].fillna('No', inplace=True)
```

```python
disease_data.isna().sum()
```

```python
disease_data['Sex'].fillna('Female', inplace=True)
disease_data['Race'].fillna('White', inplace=True)
disease_data['AgeCategory'].fillna('65-69', inplace=True)
```
